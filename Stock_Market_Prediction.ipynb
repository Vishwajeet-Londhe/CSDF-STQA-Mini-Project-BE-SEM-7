```python
df = pd.read_csv("20MICRONS.BO.csv")

df.head()
```

```python
print(df.info())
print(df.isnull().sum())
df.describe()
```

```python
df = df.dropna()
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
# 5Ô∏è‚É£ FEATURE ENGINEERING

# Daily return %
df['Return'] = df['Close'].pct_change()

# 5-day moving average
df['MA5'] = df['Close'].rolling(5).mean()

# 10-day moving average
df['MA10'] = df['Close'].rolling(10).mean()

# Volatility (past 5 days)
df['Volatility'] = df['Return'].rolling(5).std()

df.tail()
```

```python
# 6Ô∏è‚É£ TARGET: FUTURE RETURNS (NEXT DAY)

df['Future_Return'] = df['Close'].shift(-1).pct_change()

# Drop NA created by shift
df = df.dropna()
```

```python
# =====================================================================
# 7Ô∏è‚É£ FEATURES AND TARGET
features = ['Close', 'MA5', 'MA10', 'Volatility', 'Volume']
X = df[features]
y = df['Future_Return']
```

```python
# 8Ô∏è‚É£ TRAIN/TEST SPLIT
# =====================================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
# 9Ô∏è‚É£ BUILD MODEL
# =====================================================================
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train, y_train)
```

```python
# üîü MODEL PREDICTION
# =====================================================================
predictions = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, predictions))
print("R¬≤ Score:", r2_score(y_test, predictions))
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ VISUALIZE REAL VS PREDICTED RETURNS
# =====================================================================
plt.figure(figsize=(12,5))
plt.plot(y_test.values, label="Actual Return")
plt.plot(predictions, label="Predicted Return")
plt.title("Actual vs Predicted Future Returns")
plt.legend()
plt.show()
```

```python
# 1Ô∏è‚É£2Ô∏è‚É£ MARKET UPS & DOWNS ANALYSIS
# =====================================================================
df['Direction'] = np.where(df['Return'] > 0, 'Up', 'Down')
print(df['Direction'].value_counts())
```

```python
# 1Ô∏è‚É£3Ô∏è‚É£ PLOT UPS/DOWNS COUNT
# =====================================================================
df['Direction'].value_counts().plot(kind="bar", figsize=(6,4))
plt.title("Market Movement Days")
plt.show()
```
